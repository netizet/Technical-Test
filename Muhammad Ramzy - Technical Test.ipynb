## Technical Test
### 1. Write a query that finds the top 3 users with most active (frequency) on buying for each group 17-22 and 23 - 30

#### 17-22

```python
df3_0 = df3[df3['user_age'] >= 17]
df3_1 = df3_0[df3_0['user_age'] <= 22]
df3_1
```

```python
len(df3_1['user_id'].unique())
```

```python
buying_count_Saham = []

for i in df3_1['user_id'].unique():
    counter = 0
    for j in range(1, len(df3_1[df3_1['user_id'] == i])):
        if df3_1[df3_1['user_id'] == i].iloc[j]['Saham_invested_amount'] > df3_1[df3_1['user_id'] == i].iloc[j-1]['Saham_invested_amount']:
            counter += 1
    buying_count_Saham.append(counter)
```

```python
buying_count_Saham
```

```python
len(buying_count_Saham)
```

```python
buying_count_Pasar_Uang = []

for i in df3_1['user_id'].unique():
    counter = 0
    for j in range(1, len(df3_1[df3_1['user_id'] == i])):
        if df3_1[df3_1['user_id'] == i].iloc[j]['Pasar_Uang_invested_amount'] > df3_1[df3_1['user_id'] == i].iloc[j-1]['Pasar_Uang_invested_amount']:
            counter += 1
    buying_count_Pasar_Uang.append(counter)
```

```python
buying_count_Pasar_Uang
```

```python
len(buying_count_Pasar_Uang)
```

```python
buying_count_Pendapatan_Tetap = []

for i in df3_1['user_id'].unique():
    counter = 0
    for j in range(1, len(df3_1[df3_1['user_id'] == i])):
        if df3_1[df3_1['user_id'] == i].iloc[j]['Pendapatan_Tetap_invested_amount'] > df3_1[df3_1['user_id'] == i].iloc[j-1]['Pendapatan_Tetap_invested_amount']:
            counter += 1
    buying_count_Pendapatan_Tetap.append(counter)
```

```python
buying_count_Pendapatan_Tetap
```

```python
len(buying_count_Pendapatan_Tetap)
```

```python
buying_count_Campuran = []

for i in df3_1['user_id'].unique():
    counter = 0
    for j in range(1, len(df3_1[df3_1['user_id'] == i])):
        if df3_1[df3_1['user_id'] == i].iloc[j]['Campuran_invested_amount'] > df3_1[df3_1['user_id'] == i].iloc[j-1]['Campuran_invested_amount']:
            counter += 1
    buying_count_Campuran.append(counter)
```

```python
buying_count_Campuran
```

```python
len(buying_count_Campuran)
```

```python
d = {'user_id': df3_1['user_id'].unique(), 'buying_count_Saham': buying_count_Saham, 'buying_count_Pasar_Uang': buying_count_Pasar_Uang, 'buying_count_Pendapatan_Tetap': buying_count_Pendapatan_Tetap, 'buying_count_Campuran': buying_count_Campuran}
df4 = pd.DataFrame(data=d)
df4
```

```python
df4['buying_count'] = df4['buying_count_Saham'] + df4['buying_count_Pasar_Uang'] + df4['buying_count_Pendapatan_Tetap'] + df4['buying_count_Campuran']
```

```python
df4
```

```python
df4.sort_values(by = 'buying_count', ascending = False)
```
