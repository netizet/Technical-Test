### Write a query that finds the top 3 users with most active (frequency) on selling (Reksadana Saham Portfolio Only) who are female and income source not from "Keuntungan Bisnis" 

```python
df3_0 = df3[df3['user_gender'] == "Female"]
df3_1 = df3_0[df3_0['user_income_source'] != "Keuntungan Bisnis"]
```

```python
df3_1
```

```python
selling_count_Saham = []

for i in df3_1['user_id'].unique():
    counter = 0
    for j in range(1, len(df3_1[df3_1['user_id'] == i])):
        if df3_1[df3_1['user_id'] == i].iloc[j]['Saham_invested_amount'] < df3_1[df3_1['user_id'] == i].iloc[j-1]['Saham_invested_amount']:
            counter += 1
    selling_count_Saham.append(counter)
```

```python
d = {'user_id': df3_1['user_id'].unique(), 'selling_count_Saham': selling_count_Saham}
df4 = pd.DataFrame(data=d)
```
